```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('titanicsurvival.csv')
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
data['Sex']=data['Sex'].map({"male":0,"female":1})
data.head()
```

```python
X=data[['Pclass','Sex','Age','Fare']]
Y=data['Survived']
```

```python

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X=pd.DataFrame(sc.fit_transform(X))
X.columns=['Pclass','Sex','Age','Fare']
X
```

```python
#Check for null values
X.columns[X.isna().any()]
```

```python
#ill with mean values
X.Age = X.Age.fillna(X.Age.mean())
```

```python
#re-check
X.columns[X.isna().any()]
```

```python
#Split the dataset
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(X,Y,train_size=0.75,random_state=42)
```

```python
from sklearn.naive_bayes import BernoulliNB
model1=BernoulliNB().fit(train_x,train_y)
```

```python
pclassNo = int(input("Enter Person's Pclass number: "))
gender = int(input("Enter Person's Gender 0-male 1-female(0 or 1): "))
age = int(input("Enter Person's Age: "))
fare = float(input("Enter Person's Fare: "))
person = [[pclassNo,gender,age,fare]]
result = model1.predict(person)
print(result)

if result == 1:
  print("Person might be Survived")
else:
  print("Person might not be Survived")
```

```python
y_pred = model1.predict(test_x)
print(np.column_stack((y_pred,test_y)))
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the Model: {0}%".format(accuracy_score(test_y, y_pred)*100))
```

```python
from sklearn.naive_bayes import GaussianNB
model2 = GaussianNB()
model2.fit(train_x, train_y)
```

```python
y_pred2 = model2.predict(test_x)
```

```python

print("Accuracy of the Model: {0}%".format(accuracy_score(test_y, y_pred2)*100))
```
